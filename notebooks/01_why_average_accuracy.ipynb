# 1 - Why Average Accuracy Is Not Enough

A benchmark reports one number; a deployment decision needs to know *which conditions* cause failure. A system can be accurate on average and concentrate its errors on one consequential condition. This notebook makes that concrete.

```python
# 100 items under two conditions; failures concentrate on 'misleading'.
items = [('clear', 1)] * 78 + [('clear', 0)] * 2 \
      + [('misleading', 1)] * 8 + [('misleading', 0)] * 12
overall = sum(o for _, o in items) / len(items)
by_cond = {c: sum(o for cc, o in items if cc == c) / sum(cc == c for cc, _ in items)
           for c in ('clear', 'misleading')}
print(f'overall accuracy : {overall:.2f}')
print('by condition     :', {k: round(v, 2) for k, v in by_cond.items()})
```

The aggregate (0.86) hides that the system fails 60% of the time on misleading phrasing. The rest of the book builds the designed experiment that surfaces and attributes exactly this.

## Visualization

```python
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt, pathlib
FIGDIR = pathlib.Path("figures")
FIGDIR.mkdir(parents=True, exist_ok=True)
fig, ax = plt.subplots(figsize=(4, 3))
labels = ['overall', 'clear', 'misleading']
vals = [overall, by_cond['clear'], by_cond['misleading']]
ax.bar(labels, vals, color=['#444', '#2a8', '#c33'])
ax.axhline(overall, ls='--', c='#444', lw=0.8)
ax.set_ylim(0, 1); ax.set_ylabel('accuracy')
ax.set_title('Aggregate hides the weak condition')
fig.tight_layout(); fig.savefig(FIGDIR / 'fig_avg_accuracy.png', dpi=150)
print('wrote', FIGDIR / 'fig_avg_accuracy.png')
```
