# 6 - The Enrichment Design Space: *how* a question is presented

A **factor** changes the surface and context of a base question while leaving the ground truth invariant. Every factor is `gt_invariant: true` by construction - anything that could change the answer belongs in the base catalog, not here.

```python
import sys, pathlib
import proofloop.suite as gt
cat = gt.Catalog.load()
print(cat.summary())
```

## 40 factors in 10 groups, all ground-truth invariant

```python
from collections import Counter
print('factors:', len(cat.factors))
print('all gt_invariant:', all(f.gt_invariant for f in cat.factors.values()))
print(Counter(f.section for f in cat.factors.values()))
```

## `applies_to`: enrichment enriches the *base*
Each factor declares which base families / answer types it is meaningful for. When you resolve a suite, factors that match no selected base are dropped automatically - so a numeric-format factor never decorates a set-valued cross-reference question.

```python
# cross_reference answers set_str -> numeric_format does NOT apply
s1 = gt.resolve(cat, ['cross_reference'], ['numeric_format','clarity'], mode='embedded')
print('cross_reference  kept :', s1.factor_names)
print('cross_reference  drop :', s1.dropped_factors)
# exact_recall answers float -> numeric_format applies
s2 = gt.resolve(cat, ['exact_recall'], ['numeric_format','clarity'], mode='embedded')
print('exact_recall     kept :', s2.factor_names)
```

## Factor groups are reusable bundles
The catalog ships the same groups as the knowlytix factor catalog (`quick_screen`, `comprehensive`, `full`, ...).

```python
for g in ('quick_screen','adversarial','comprehensive'):
    print(f'{g:14s} {cat.factor_groups[g]}')
```

## Application level vocabulary
The catalog is domain-neutral. An application supplies its own level labels via `level_overrides` (kept out of the general catalog). For example the complaint agent renames `entity_aliasing` levels to `canonical / alias / abbreviated` - applied at `compose` time, shown in notebook 3.

## Self-check

```python
assert len(cat.factors) == 40
assert all(f.gt_invariant for f in cat.factors.values())
assert 'numeric_format' not in s1.factor_names   # dropped for set_str base
assert 'numeric_format' in s2.factor_names        # kept for float base
print('OK - enrichment catalog')
```
